# Principal Component Analysis on Hand Written Digit Data Set

### Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

### The Data

```python
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
type(digit)
```

```python
digit.keys()
```

```python
df = pd.DataFrame(digit['data'],columns=digit['feature_names'])
```

```python
df.head()
```

# PCA Visualization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=digit['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
sns.scatterplot(x = x_pca[:,0],y = x_pca[:,1])
```

```python
sns.scatterplot(x = x_pca[:,0],y = x_pca[:,1],hue=digit['target'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = x_pca[:,0], y = x_pca[:,1],hue=digit['target'],palette='Set1')
plt.legend(loc=(1.05,0))
```
